```python

# importing the required module
import matplotlib.pyplot as plt
```

```python
def cal_battery_life(t_stand,t_data,t_poll,t_total):
    I_stand = 0.001 #mA
    I_data = 11.89 #mA
    I_poll = 12.87 #mA
    batttery_capacity = 240; #unit mA
    derating = 0.85;

    total_consumation = ((I_stand*t_stand)+(I_data*t_data)+(I_poll*t_poll))/(t_total);

    battery_life_sec = batttery_capacity/total_consumation;

    battery_life_years = (battery_life_sec / 8760)*derating;

    return  battery_life_years
    
```

```python
def is_what_percent_of(num_a, num_b):
    return (num_a / num_b) * 100
```

```python
def plot_final_graph(msgs,b,b1,b2,b3):
    # Plotting both the curves simultaneously
    plt.semilogy(msgs, b, color='r', linewidth=2, marker ='*',label='Polling Interval 1200 sec')
    plt.semilogy(msgs, b1, color='g', linewidth=2, marker ='+',label='Polling Interval 500 sec')
    plt.semilogy(msgs, b2, color='b', linewidth=2, marker ='o',label='Polling Interval 250 sec')
    plt.semilogy(msgs, b3, color='y', linewidth=2, marker ='x',label='Polling Interval 60 sec')
    plt.legend()
    # naming the x axis
    plt.xlabel('Data Messages per hour')
    # naming the y axis
    plt.ylabel('Battery Life in years')
  
    # giving a title to my graph
    plt.title('Battery Life Estimation')
    
    
    plt.grid()
  
    # function to show the plot
    plt.show()
```

```python
def plot_graph(b,msgs):
    #plt.yscale("log")
    # plotting the points 
    plt.semilogy(b, msgs)
  
    # naming the x axis
    plt.xlabel('Data Messages per hour')
    # naming the y axis
    plt.ylabel('Battery Life in years')
  
    # giving a title to my graph
    plt.title('Battery Life Estimation')
  
    # function to show the plot
    plt.show()
```

```python
def main():
    print("Program for calculation of Battery Life");
    print("Battery Life Calculation for Humidity and Temperature Sensor Node C1310");
    print(" Power Source Model = CR2032 Lithium-ion coin cell battery ");
    print("Normal Load parameters");
    print("Battery capacity = 240 mA")
    print("I_Standby (macroA)= 1.0 macroA");
    print("T_standby = 1200 sec");
    print("I_data = 11.89 mA");
    print("T_data = n data packets events/hr * time to send single data packets =165.68");
    print("n= 12 data packets, time to send single data packet = 41.42ms");
    print("I poll = 12.87mA");
    print("T poll = 15 ms")
    print("Derating factor = 0.85");
    print("Polling interval = 1200s");
    #calculate normal load   


    t_poll = 15/1000; #15 ms 
    t_total = 1200; # sec configurable
    t_stand = 1200;
    t_single_data = 41.42; #ms
    n = 12 # 12 packets per hour - configurable
    
    t_data = (n*t_single_data*1200)/3600; #ms
    t_data = t_data/1000 # sec
    print("t_data = "+str(t_data)+" sec");
    
    battery_life_years = cal_battery_life(t_stand,t_data,t_poll,t_total)
    print("\n Estimated battery life for normal load 12 msgs = "+str(round(battery_life_years,2))+" years");
    
   

    print("Change only msgs")
    msgs=[180,360,540,720,900,1080,1260,1440,1620,1800,1980,2160,2340,2520,2700,2880,3060,3240,3420,3600]
    b=[]
    
    for n in msgs:
        print("\n Change msgs to "+str(n)+" = "+str(is_what_percent_of(n,3600))+"%")
        t_data = (n*t_single_data*1200)/3600; #ms
        t_data = t_data/1000 # sec
        print("  t_data = "+str(t_data)+" sec");
        battery_life_years = cal_battery_life(t_stand,t_data,t_poll,t_total)
        print(" Estimated battery life load = "+str(round(battery_life_years,3))+" years");
        b.append(round(battery_life_years,3))   
    
    if (battery_life_years < 1 ):
        days = battery_life_years * 365
        print("\nbattery life =",b," \n In Days = ",days) 
    else:
        print("\nbattery life =",b)
    plot_graph(msgs,b)
    
    
    print("Change msgs and polling interval to 500")
    t_total = 500
    b1=[]
    for n in msgs:
        print("\n Change msgs to "+str(n)+" = "+str(is_what_percent_of(n,3600))+"%")
        t_data = (n*t_single_data*1200)/3600; #ms
        t_data = t_data/1000 # sec
        print("  t_data = "+str(t_data)+" sec");
        battery_life_years = cal_battery_life(t_stand,t_data,t_poll,t_total)
        print(" Estimated battery life load = "+str(round(battery_life_years,3))+" years");
        b1.append(round(battery_life_years,3))   
    if (battery_life_years < 1 ):
        days = battery_life_years * 365
        print("\nbattery life =",b1," \n In Days = ",days) 
    else:
        print("\nbattery life =",b1)  
    plot_graph(msgs,b1)
    
    
    print("Change msgs and polling interval to 250")
    t_total = 250
    b2=[]
    for n in msgs:
        print("\n Change msgs to "+str(n)+" = "+str(is_what_percent_of(n,3600))+"%")
        t_data = (n*t_single_data*1200)/3600; #ms
        t_data = t_data/1000 # sec
        print("  t_data = "+str(t_data)+" sec");
        battery_life_years = cal_battery_life(t_stand,t_data,t_poll,t_total)
        print(" Estimated battery life load = "+str(round(battery_life_years,3))+" years");
        b2.append(round(battery_life_years,3))   
    if (battery_life_years < 1 ):
        days = battery_life_years * 365
        print("\nbattery life =",b2," \n In Days = ",days) 
    else:
        print("\nbattery life =",b2) 
    plot_graph(msgs,b2)
    
    print("Change msgs and polling interval to 60")
    t_total = 60
    b3=[]
    for n in msgs:
        print("\n Change msgs to "+str(n)+" = "+str(is_what_percent_of(n,3600))+"%")
        t_data = (n*t_single_data*1200)/3600; #ms
        t_data = t_data/1000 # sec
        print("  t_data = "+str(t_data)+" sec");
        battery_life_years = cal_battery_life(t_stand,t_data,t_poll,t_total)
        print(" Estimated battery life load = "+str(round(battery_life_years,3))+" years");
        b3.append(round(battery_life_years,3))   
    if (battery_life_years < 1 ):
        days = battery_life_years * 365
        print("\nbattery life =",b3," \n In Days = ",days) 
    else:
        print("\nbattery life =",b3) 
    plot_graph(msgs,b3)
    
    plot_final_graph(msgs,b,b1,b2,b3)
    
    
if __name__ == "__main__":
    main()
```
